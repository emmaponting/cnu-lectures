```python
import math
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0, 1, 10) #make an array
c = lambda n: np.cos(n*x)
s = lambda n: np.sin(n*x)
```

```python
plt.plot(x, c(1))
plt.plot(x, s(1))
plt.plot(x, c(2))
plt.plot(x, s(2))
```

```python
#fig, ax = plt.subplots(2, 4, sharex = False, sharey = False)
#2 rows and 4 columns

fig, ax = plt.subplots(2, 4, sharex = True, sharey = True)

for j in range(4):
    ax[0,j].plot(x, c(j), color = 'blue', lw = 2.5)
    ax[1,j].plot(x, s(j), color = 'blue', lw = 2.5)

```
